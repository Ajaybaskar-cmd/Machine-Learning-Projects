### 0-> Non Fraudlent
### 1 ->    Fraulent

This dataset is the unbalanced data so Upsampling the data counts

```python
non_fraudlent = df[df['Class']==0]
fraudlent = df[df['Class']==1]
```

### Upsampling the fraudlent data

```python
fraudlent.shape
```

```python
upsample_data = fraudlent.sample(n= len(non_fraudlent),replace=True)
upsample_data.shape
```

## Concatenate the two dataframe for 
upsample_data,non_fraudlent

```python
cred_df = pd.concat([non_fraudlent,upsample_data],axis = 0)
cred_df.head()
```

```python
# value counts after upsampling
cred_df['Class'].value_counts()
```

### Splitting the dataset target and features

```python
x = cred_df.drop(columns = 'Class',axis = 1)
y = cred_df['Class']
```

```python
scale = StandardScaler()

X = scale.fit_transform(x)
```

```python
print(X[:5])
```

```python
print(y.head())
```

### Split the training and test data

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,stratify = y,random_state = 42)
```

```python
print(X.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

# Model Training

#### Logistic Regression

```python
model = LogisticRegression()
model
```

```python
# training the logistic regression model on training data
model.fit(x_train,y_train)
```

# Model Evaluation

```python
# accuaracy score on training data
train_data_pred = model.predict(x_train)
train_data_accuracy = accuracy_score(train_data_pred,y_train)
```

```python
print("Accuracy Score on the train data :",train_data_accuracy)
```

```python
print(classification_report(y_train,train_data_pred))
```

```python
# accuracy score on test data
test_data_pred = model.predict(x_test)
test_data_accuracy = accuracy_score(test_data_pred,y_test)
```

```python
print("Accuracy Score on the test data : ",test_data_accuracy)
```

```python
print(classification_report(y_test,test_data_pred))
```

```python
# for loop using two algorithms
models = {"LogisticRegression" : LogisticRegression(),
          "DecisionTreeClassifier":DecisionTreeClassifier()
}

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(x_train,y_train)
    
    #train data prediction
    train_data_pred = model.predict(x_train)
    
    #train data model evaluation
    train_data_accu = accuracy_score(train_data_pred,y_train)
    
    #test data prediction
    test_data_pred = model.predict(x_test)
    
    # test data accuracy
    test_data_accu = accuracy_score(test_data_pred,y_test)
    
    # print the accuracy score 
    print(list(models.keys())[i])
    print("Train Data Accuracy : ",train_data_accu)
    print("Test Data Accuracy : ",test_data_accu)
```
